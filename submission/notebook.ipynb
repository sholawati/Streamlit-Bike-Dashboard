## 1. Load Dataset

```python
import pandas as pd
df = pd.read_csv('data/day.csv')
```

## 2. Bersihkan Data (Preprocessing)
- Ubah kolom tanggal jadi datetime
- Mapping season, weathersit
- Rename kolom biar mudah dibaca

```python
df['dteday'] = pd.to_datetime(df['dteday'])
df.rename(columns={
    'dteday': 'date',
    'yr': 'year',
    'hum': 'humidity',
    'cnt': 'total_rentals',
    'atemp': 'feels_like',
    'weathersit': 'weather'
}, inplace=True)
df['season'] = df['season'].map({1:'Spring', 2:'Summer', 3:'Fall', 4:'Winter'})
df['weather'] = df['weather'].map({1:'Clear', 2:'Mist/Cloudy', 3:'Light Rain/Snow', 4:'Heavy Rain/Snow'})
df['year'] = df['year'].map({0: 2011, 1: 2012})
```

## 3. Analisis Data (EDA)
- Tren penyewaan sepeda per bulan/tahun
- Pengaruh musim/cuaca terhadap penyewaan
- Insight tambahan seperti pola musiman

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Contoh grafik
sns.lineplot(x='date', y='total_rentals', data=df)
plt.title("Tren Penyewaan Sepeda")
plt.show()
```

## 4. Simpan hasil akhir jadi main_data.csv

```python
df.to_csv('dashboard/main_data.csv', index=False)
```
